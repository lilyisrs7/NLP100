```python
from pickle import load
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

with open('country_vecs.pkl', 'rb') as f:
  country_vecs = load(f)
annotations = list(country_vecs.keys())
tsne = TSNE(n_components=2, random_state=0).fit_transform(np.array(list(country_vecs.values())))
plt.scatter(*zip(*tsne))
for i, label in enumerate(annotations):
    plt.annotate(label, tsne[i], fontsize=7)
```
