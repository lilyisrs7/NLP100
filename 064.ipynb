```python
import os

with open('questions-words.csv') as f:
  writefile = '64-questions-words.csv'
  if os.path.isfile(writefile):
    mode = 'w'
  else:
    mode = 'x'
  with open(writefile, mode=mode) as fw:
    fw.write('category 1 2 3 4\n')
    for line in f.readlines():
      if line.startswith(':'):
        category = line.split()[1]
      else:
        fw.write(category + ' ' + line)
```

```python
import pandas as pd

df = pd.read_csv('64-questions-words.csv', sep=' ')
df.head()
```

```python
len(df)
```

```python
import numpy as np
from gensim.models import KeyedVectors

model = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
def func(pos1, pos2, neg):
  return model.most_similar([pos1, pos2], neg, topn=1)[0]

len_df_i = len(df) // 8 # 2443
for i in range(8):
  df_i = df.iloc[len_df_i*i:len_df_i*(i+1)]
  df_i['word'], df_i['similarity'] = np.vectorize(func)(df_i['2'], df_i['3'], df_i['1'])
  df_i.to_csv('64-questions-words' + str(i) + '.csv')
```

```python
df_list = [pd.read_csv('64-questions-words' + str(i) + '.csv') for i in range(8)]
df = pd.concat(df_list).drop(columns=['Unnamed: 0'])
df.to_csv('64-questions-words.csv', index=False)
```
