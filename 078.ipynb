```python
from joblib import load

X_train = load('X_train.joblib')
y_train = load('y_train.joblib')
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import time

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
train_dataset = TensorDataset(torch.tensor(X_train), torch.tensor(y_train))

epochs = 100
lr = 0.2
momentum = 0.9
criterion = nn.CrossEntropyLoss()

for batch_size in [2**i for i in range(8)]:
  model = nn.Linear(in_features=300, out_features=4, dtype=torch.float64)
  model = model.to(device)
  optimizer = optim.SGD(model.parameters(), lr=lr, momentum=momentum)
  train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
  print('batch size:', batch_size)
  for epoch in range(epochs):
    acc = 0
    start = time.time()
    model.train()
    for X_train_, y_train_ in train_loader:
      X_train_ = X_train_.to(device)
      y_train_ = y_train_.to(device)
      optimizer.zero_grad()
      y_pred = model(X_train_)
      loss = criterion(y_pred, y_train_)
      loss.backward()
      optimizer.step()
      acc += sum(torch.max(y_pred, 1)[1] == y_train_).item()
      stop = time.time()
    print('epoch:', epoch+1, 'elapsed time(s):', stop-start, 'acc:', acc / len(y_train))
```
