```python
from pickle import load

with open('country_vecs.pkl', 'rb') as f:
  country_vecs = load(f)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

plt.figure(figsize=(24, 12))
link = linkage(list(country_vecs.values()), method='ward')
dendrogram(link, labels=list(country_vecs.keys()))
plt.xticks(fontsize=12)
plt.show()
```
