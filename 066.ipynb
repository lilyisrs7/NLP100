```python
import pandas as pd

df = pd.read_csv('combined.csv')
df.head()
```

```python
import numpy as np
from gensim.models import KeyedVectors

model = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
df.sort_values(by=['Human (mean)'], ascending=False, inplace=True, ignore_index=True)
df['human_rank'] = [i for i in range(1, len(df)+1)]
def func(w1, w2):
  return model.similarity(w1, w2)
df['wordvec'] = np.vectorize(func)(df['Word 1'], df['Word 2'])
df.sort_values(by=['wordvec'], ascending=False, inplace=True, ignore_index=True)
df['wordvec_rank'] = [i for i in range(1, len(df)+1)]
df.head()
```

```python
from scipy.stats import spearmanr

spearmanr(df['human_rank'], df['wordvec_rank']).statistic
```
